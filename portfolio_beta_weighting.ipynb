```python
import datetime as dt
import pandas as pd
import numpy as np
from scipy import stats
from pandas_datareader import data as pdr
import yfinance as yf
```

```python
start= dt.datetime(2022,2,1)
end=dt.datetime(2022,2,5)
```

```python
stocklist=['TITAN', "BPCL", "LT", "TATASTEEL", "SBILIFE", "HDFCLIFE", "RELIANCE", "BHARTIARTL", "SUNPHARMA", "NTPC", "POWERGRID", "INDUSINDBK",
           "HEROMOTOCO", "HINDUNILVR", "COALINDIA", "CIPLA", "ULTRACEMCO", "HDFCBANK", "JSWSTEEL", "HINDALCO", "NESTLEIND", "TATACONSUM",
           "BRITANNIA", "HDFC", "ONGC", "DRREDDY", 'ITC', "DIVISLAB", 'MARUTI', "BAJAJ-AUTO", 'TCS', "GRASIM", 'ASIANPAINT', 'INFY',
           'APOLLOHOSP', 'ICICIBANK', 'WIPRO', 'TATAMOTORS', 'EICHERMOT', 'KOTAKBANK', 'M&M', 'SBIN', 'BAJAJFINSV', 'AXISBANK',
           'HCLTECH', 'BAJFINANCE', 'UPL', 'TECHM', 'ADANIPORTS', 'ADANIENT']
stocks =['^NSEI']+[i+'.NS' for i in stocklist]
print((stocks))
```

```python
yf.pdr_override()
df=pdr.get_data_yahoo(stocks,start,end)
log_returns=np.log(df.Close/df.Close.shift(1)).dropna()
log_returns
```

CALCULATING BETA USING THE FORMULA: (COV(MARKET,STOCK)/VAR(MARKET)

```python
def calc_beta(dl):
  np_array = dl.values
  # market index is the last column
  m = np_array[:,-1]
  beta = []
  for ind,col in enumerate(df):
    if ind <50:
      # stock returns indexed by ind
      s = np_array[:,ind]
      # calculate covariance matrix between stock and market
      covariance = np.cov(s,m)
      beta.append(covariance[0,1]/covariance[1,1])
  return pd.Series(beta,dl.columns[:-1],name='Beta')

  
```

```python
calc_beta(log_returns)
```

CALCULATING BETA USING REGRESSION

```python
def regression_beta(dl):
  np_array = dl.values
  # market index is the last column
  m = np_array[:,-1]
  beta = []
  for ind,col in enumerate(df):
    if ind <50:
      # stock returns indexed by ind
      s = np_array[:,ind]
      beta.append(stats.linregress(m,s)[0])
  return pd.Series(beta,dl.columns[:-1],name='Beta')
```

```python
beta = regression_beta(log_returns)
```
